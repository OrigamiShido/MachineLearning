# 使用KNN算法实现约会网站的配对效果

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.preprocessing import minmax_scale
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix

#  读取datingtestset2数据

data = pd.read_csv("C:/Users/Admin/Desktop/WHU study/programming/python/MachineLearning/exp1/datingTestSet2.txt", sep='\t', header=None)
data.head()# line0:planemile,line1:video game percentage,line2:icecreamliter,line3:label

# 归一化
scaler=MinMaxScaler()
scaler=scaler.fit(data[[0,1,2]])
result=scaler.transform(data[[0,1,2]])
print(result)

# 划分数据集
X_train, X_test, y_train, y_test = train_test_split(result, data[3], test_size=0.2, random_state=0)

# 训练模型
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# 预测
y_pred = knn.predict(X_test)

# 评估
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

# 预测
print(knn.predict([[0.1,0.2,0.3]]))

# 画图
print(data[[1,2]])

fig=plt.figure()
ax=fig.add_subplot(111)
ax.scatter(data[1],data[2])
plt.show()

```
